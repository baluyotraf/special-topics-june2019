```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_NIGHT_PATH = 'piova_night.png'
IMAGE_DAY_PATH = 'piova_day.png'
```

```python
img_night = cv2.imread(IMAGE_NIGHT_PATH, cv2.IMREAD_COLOR) / 255.0
img_night = img_night[..., ::-1]
img_day = cv2.imread(IMAGE_DAY_PATH, cv2.IMREAD_COLOR) / 255.0
img_day = img_day[..., ::-1]
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_day)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_night)
plt.axis('off')

plt.show()
```

```python
abs_difference = np.abs(img_day - img_night)
abs_difference = np.clip(abs_difference, 0, 1)
```

```python
channels = ['Red', 'Blue', 'Green']
plt.figure(figsize=(20, 20))
for i, c in enumerate(range(0, 3), 1):
    plt.subplot(1, 3, i)
    plt.title(f'Absolute Difference in {channels[c]} Channel')
    plt.imshow(abs_difference[..., c], cmap='gray')
    plt.axis('off')
plt.show()
```
