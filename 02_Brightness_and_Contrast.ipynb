```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_COLOR)
# Opencv reads image in BRG and not in RGB so we convert
img = img[..., ::-1]
# Convert the numbers to an floating point intensity array
img = img / 255.0
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
brightness_scale = 1.5
bright_img = np.clip(img * brightness_scale, 0, 1)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Bright Image')
plt.imshow(bright_img)
plt.axis('off')

plt.show()
```

```python
constrast_scale = 1.2
constrast_img = img**constrast_scale
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Constrast Image')
plt.imshow(constrast_img)
plt.axis('off')

plt.show()
```
