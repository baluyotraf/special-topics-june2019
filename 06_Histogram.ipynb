```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
from skimage import util
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def plot_image_histogram(image):
    image = util.img_as_ubyte(image)
    pixel, count = np.unique(image, return_counts=True)
    all_pixel = np.arange(0, 256, dtype=np.uint8)
    all_count = np.zeros(256, dtype=np.uint8)
    all_count[pixel] = count
    plt.bar(all_pixel, all_count)
    plt.show()
    
plot_image_histogram(img)
```

```python
global_histogram = exposure.rescale_intensity(img, in_range=(0, 150))
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(global_histogram, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plot_image_histogram(global_histogram)
```

```python
equalization_histogram = exposure.equalize_hist(img)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalization_histogram, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plot_image_histogram(equalization_histogram)
```

```python
adaptive_histogram = exposure.equalize_adapthist(img, 200)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(adaptive_histogram, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plot_image_histogram(adaptive_histogram)
```
