```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import segmentation
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_COLOR)[..., ::-1]
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
felzenszwalb_1 = segmentation.felzenszwalb(img, scale=1)
felzenszwalb_1_img = segmentation.mark_boundaries(img, felzenszwalb_1)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(felzenszwalb_1_img)
plt.axis('off')
plt.show()
```

```python
felzenszwalb_2 = segmentation.felzenszwalb(img, scale=100)
felzenszwalb_2_img = segmentation.mark_boundaries(img, felzenszwalb_2)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(felzenszwalb_2_img)
plt.axis('off')
plt.show()
```

```python
quickshift = segmentation.quickshift(img)
quickshift_img = segmentation.mark_boundaries(img, quickshift)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(quickshift_img)
plt.axis('off')
plt.show()
```

```python
slic = segmentation.slic(img, n_segments=1000)
slic_img = segmentation.mark_boundaries(img, slic)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(slic_img)
plt.axis('off')
plt.show()
```
