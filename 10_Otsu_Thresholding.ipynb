```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
threshold, mask_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.imshow(mask_img, cmap='gray')
plt.axis('off')
plt.show()
```
