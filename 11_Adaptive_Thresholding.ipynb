```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
mask_img_1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 75, 0)
```

```python
plt.imshow(mask_img_1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
mask_img_2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 75, 0)
```

```python
plt.imshow(mask_img_2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
mask_img_3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 75, -40)
```

```python
plt.imshow(mask_img_3, cmap='gray')
plt.axis('off')
plt.show()
```
