```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = 'piova_night.png'
```

```python
img = cv2.imread(IMAGE_PATH, cv2.IMREAD_COLOR)
# Opencv reads image in BRG and not in RGB so we convert
img = img[..., ::-1]
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
def quantize(image, n_bits):
    image = image >> (8 - n_bits)
    image = image * 255.0 / (2**n_bits-1)
    return image.astype(np.uint8)
```

```python
quantizations = [quantize(img, n_bits) for n_bits in range(1, 8+1)]
```

```python
plt.figure(figsize=(20, 7))
for i, q in enumerate(quantizations, 1):
    plt.subplot(2, 4, i)
    plt.title(f'Quantization using {2**i} bits')
    plt.axis('off')
    plt.imshow(q)
plt.show()
```
